# 그래프 그리기 및 데이터 처리를 위한 패키지들

```python
import numpy as np                 # 데이터 처리를 위한 모듈
import matplotlib.pyplot as plt    # 그래프를 그리는 모듈
```

# 내가 정의한 함수: $f(x) = 2x^7 - x^5 + 4x^4 - 2x^3 + 5x^2 + 3$ 
\* 학번에서 착안(2 0 1 4 2 5 0 3)

```python
def myFunction(x):
    f = 2*x**7 - x**5 + 4*x**4 - 2*x**3 + 5*x**2 + 3
    return f
```

# 내가 정의한 함수의 도함수: $f'(x) = 14x^6 - 5x^4 + 16x^3 - 6x^2 + 10x$

```python
def myDerivativeFunction(x):
    df = 14*x**6 - 5*x**4 + 16*x**3 - 6*x**2 + 10*x
    return df
```

# 함수의 정의역: $x = [-2.5 : 0.01 : 2.5]$

```python
x = np.arange(-2.5, 2.5, 0.01)
```

# 그래프 계산하기

```python
f = myFunction(x)
df = myDerivativeFunction(x)
```

# 그래프 그리기

```python
plt.figure(1)
plt.plot(x, f, 'b', label="function")
plt.plot(x, df, 'r', label="derivative")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

# 내가 정의한 함수에서 3개의 점 선택 (x = -2 , 1, 2)

```python
pointX = np.array([-2, 1, 2])            # x좌표   -2,  1,   2
pointY = np.array([-121.0, 11.0, 295.0]) # y좌표 f(-2),f(1),f(2)                                         
```

```python
plt.figure(1)
plt.plot(x, f, 'b', label="function")
plt.scatter(pointX, pointY)              # 점 표시 
plt.show()                               # (-2, f(-2)) , (1, f(1)), (2, f(2))
```

# $x = -2$ 점에서의 first-order Taylor expansion
## $\hat{f}(x) = f(-2) + f'(-2)(x - (-2))$
## $\hat{f}(x) = -121 + 644 (x + 2)$
## $\hat{f}(x) = 664x + 1167$

```python
def taylorExpansion0(x):   # x = -2 에서의 taylor expansion 정의
    te0 = 664*x + 1167
    return te0

te0 = taylorExpansion0(x)  # 그래프 계산 
```

# 정의한 함수와 $x = -2$ 에서의 taylor expansion 그래프

```python
plt.figure(1)
plt.plot(x, f, 'b', label="function")
plt.plot(x, te0, 'r', label="taylor expansion at -2")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

# x값 범위 (-2.5 ~ -1.5) 줄인 그래프

```python
plt.figure(1)
plt.plot(x, f, 'b', label="function")
plt.plot(x, te0, 'r', label="taylor expansion at -2")

plt.xlim(-2.5, -1.5)       # 더 자세히 관찰 위해 x값 표시 제한

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

# x값 범위 (-2.3 ~ -1.7) 줄인 그래프

```python
plt.figure(1)
plt.plot(x, f, 'b', label="function")
plt.plot(x, te0, 'r', label="taylor expansion at -2")

plt.xlim(-2.3, -1.7)       # 더 자세히 관찰 위해 x값 표시 제한

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

# $x = 1$ 점에서의 first-order Taylor expansion
## $\hat{f}(x) = f(1) + f'(1)(x - 1)$
## $\hat{f}(x) = 11 + 29 (x - 1)$
## $\hat{f}(x) = 29x - 18$

```python
def taylorExpansion1(x):   # x = 1 에서의 taylor expansion 정의
    te1 = 29*x - 18
    return te1

te1 = taylorExpansion1(x)  # 그래프 계산 
```
